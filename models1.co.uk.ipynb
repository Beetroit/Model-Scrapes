```python
import requests, csv, os, re
from bs4 import BeautifulSoup
models_details = list()
main_dict=dict()
basedir = 'models1.co.uk'
base_url = 'models1.co.uk'
url = 'https://www.models1.co.uk/women/main/'
```

```python
soup = BeautifulSoup(requests.get(url).content, 'html5lib')
```

```python
urls = [link['href'] for link in soup.find('div' ,attrs={"class":'categoriesInSelection'}).findAll('a') ]
soups = [BeautifulSoup(requests.get(url).content, 'html5lib') for url in urls]
soup_dict = dict(zip(urls, soups))
model_page = [[link,model.div['onclick']] for link, soup in soup_dict.items() for model in soup.findAll('div', class_='modelsElement')]
```

```python
def download_pics(url, k):
    soup = BeautifulSoup(requests.get(url).content, 'html5lib')
    pics = [pic.img['data-lazy'] for pic in soup.findAll('div', class_='portfolioElement')[:-1]]
    for pic in pics:
        print(pic)
        path = f"{basedir}/{k}/{pic.split('/')[-1]}" if "?" not in pic else f"{basedir}/{k}/{pic.split('/')[-1].split('?')[0]}"
        # if not os.path.exists(path):
        #     with open(path, 'wb') as f:
        #         f.write(requests.get(pic).content)

def get_model_link(text):
    regex = re.compile(r"\((.*?)\)")
    matches = regex.search(text[1]).group(0) 
    text[1]="-".join(matches.strip("()'").replace("'",'').split(','))
    return (''.join(text))

def list_to_csv(lst) -> None:
    with open(f"{basedir}/models_info.csv",'w') as f:
        c = csv.DictWriter(f, ["NAME","URL", 'HEIGHT', 'BUST', 'WAIST', 'HIPS',  'SHOE', 'EYES','HAIR'])
        c.writeheader()
        for row in lst:
            c.writerow(row)


def parse(url):
    soup = BeautifulSoup(requests.get(url).content, 'html5lib')
    raw = ["-".join(feat.text.split(" ")[:2]) for feat in soup.find('div', class_='bookMeasurements').findAll('li')]

    features = [feat.split('-')[0].upper() for feat in raw]
    values = [val.split('-')[1] for val in raw]
    return dict(zip(features,values))
```

```python

models = [get_model_link(lst) for lst in model_page]
names = [' '.join([name.title() for name in model.split('-')[1:]]) for model in models]
models_dict = dict(zip(names,models))
models_dict
```

```python

for k,v in models_dict.items():
    os.makedirs(f"{basedir}/{k}", exist_ok=True)
    print(k)
    feat_dict={'NAME':k, "URL":v}
    feat_dict.update(parse(v))
    download_pics(v,k)
    models_details.append(feat_dict)
```
