```python
import requests, os, csv
from bs4 import BeautifulSoup
url = 'https://cmmodels.com'
```

```python
def get_models():
    try:
        os.mkdir(url.split('/')[-1])
    except:
        pass
    r = requests.get(url+'/women/')
    soup = BeautifulSoup(r.content,'html5lib')

    href_list = soup.find('div', attrs={'id':'av-masonry-2'}).findAll('a')

    model_link = [i['data-av-masonry-custom-link'] for i in href_list]
    model_names = [i['title'].split('<br')[0] for i in href_list]

    models_dict =dict(zip(model_names,model_link))
    for i in models_dict.keys():
        if '\t' in i:
            i = i.replace('\t',' ')
        os.makedirs(f"{url.split('/')[-1]}/{i.split('/')[-1]}",exist_ok=True)
    print(models_dict)
    return models_dict
get_models()
```

```python
def get_jpegs(models_dict : dict):
    models_info = list()
    for name,link in models_dict.items():
        model_info = dict()
        model_info['Name'] = name.replace('\t', ' ') if '\t' in name else name
        r2 = requests.get(url+link)
        model_info['Url'] = url+r2.url
        print(model_info.get('Url'))
        soup2 = BeautifulSoup(r2.content, 'html5lib')
        
        details = soup2.findAll('td')
        for info in details:
            info = info.text.split('\n') 
            model_info[info[0]]=info[1]
        for i in soup2.findAll('a'):
            if ".jpg" in i['href'] and 'https://' in i['href']:
                jpeg_url = i.get('href')
                alt_text = i.get('href').split('/')[-1]
                
                # write image bin to file
                if not os.path.exists(f"./{url.split('/')[-1]}/{alt_text.split('-')[0]}/{alt_text}"):
                    try:
                        with open(f"./{url.split('/')[-1]}/{alt_text.split('-')[0]}/{alt_text}", 'wb') as f:
                            f.write(requests.get(jpeg_url).content)
                    except FileNotFoundError:
                        os.system(f"mkdir ./{url.split('/')[-1]}/{alt_text.split('-')[0]}/{alt_text}")
                        with open(f"./{url.split('/')[-1]}/{model_info['Name']}/{alt_text}", 'wb') as f:
                            f.write(requests.get(jpeg_url).content)
        print(model_info)
        models_info.append(model_info)
    return models_info
                
```

```python
def list_to_csv(lst):
    with open("./cmmodels.com/models_info.csv",'w') as f:
        c = csv.DictWriter(f, ['Name','Url','Height','Dress','Shoes','Breast','Waist','Hips'])
        c.writeheader()
        for row in lst:
            c.writerow(row)
```

```python
r = requests.get(url)
print(r.status_code)
```

```python

models = get_models()

models_info = get_jpegs(models)
```

```python

list_to_csv(models_info)
```
